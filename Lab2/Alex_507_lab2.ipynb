```python
text = '''Machine Learning (ML) is fascinating! It’s not just about algorithms;
it’s about data, too. Don’t you think the impact of AI on industries—healthcare,
finance, education, etc.—is revolutionary? From predictive modeling to natural language processing (NLP), 
AI is transforming the way we live and work. However, it’s not always perfect. 
Models don’t always predict accurately when faced with biased or incomplete data. 
That’s where data cleaning, feature engineering, and regularization techniques come into play! 
For more details or collaboration opportunities, feel free to contact me at ai.expert@example.com or call +91-9876543210. 
By the way, my Date of Birth is 27-March-2000. Special characters like %, $, and # are often encountered during data preprocessing, 
especially when dealing with financial datasets. And let’s not forget the importance of ethical AI—bias mitigation and fairness are key! 
But don’t worry, with the right techniques and a robust dataset, even noisy data can produce valuable insights and drive innovation forward.'''
```

```python
#downloading neccessary files
import nltk
nltk.download('averaged_perceptron_tagger')
nltk.download('punkt')

nltk.download('stopwords')
```

```python
#import the libraries
import nltk
from nltk.tokenize import word_tokenize

# Tokenize the text
tokens = word_tokenize(text)

# # Count unique tokens
unique_tokens = set(tokens)
unique_token_count = len(unique_tokens)

print(f"Number of unique tokens: {unique_token_count}")
```

```python
#Count the number of punctuation characters and remove punctuations.
import string

# Count punctuation characters
punctuation_count = sum(1 for char in text if char in string.punctuation)
print(f"Number of punctuation characters: {punctuation_count}")

# Remove punctuation
text_without_punctuation = ''.join(char for char in text if char not in string.punctuation)
print("\nText without punctuation:\n")
print(text_without_punctuation)
```

```python
#Plot the distribution of stopwords used in the text using barplot.
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import matplotlib.pyplot as plt
from collections import Counter


# Tokenize the text
tokens = word_tokenize(text.lower())

# Get English stopwords
stop_words = set(stopwords.words('english'))

# Filter out stopwords from the tokens
stopword_tokens = [word for word in tokens if word in stop_words]

# Count the frequency of each stopword
stopword_freq = Counter(stopword_tokens)

# Plotting the stopword distribution
plt.figure(figsize=(10, 6))
plt.bar(stopword_freq.keys(), stopword_freq.values(), color='skyblue')
plt.xlabel('Stopwords')
plt.ylabel('Frequency')
plt.title('Distribution of Stopwords in the Text')
plt.xticks(rotation=45)
plt.show()
```

```python
#Remove the stopwords from the text.

# Remove stopwords from the tokens
filtered_tokens = [word for word in tokens if word not in stop_words]

# Join the filtered tokens back into a single string
text_without_stopwords = ' '.join(filtered_tokens)

print("Text without stopwords:\n")
print(text_without_stopwords)
```

```python
#Plot the distribution of each POS Tag using a barplot.
from nltk import pos_tag
# POS tagging
pos_tags = pos_tag(tokens)

# Count POS tags
pos_counts = Counter(tag for word, tag in pos_tags)

# Plotting the POS distribution
plt.figure(figsize=(12, 6))
plt.bar(pos_counts.keys(), pos_counts.values(), color='lightgreen')
plt.xlabel('POS Tags')
plt.ylabel('Frequency')
plt.title('Distribution of POS Tags in the Text')
plt.xticks(rotation=45)
plt.show()
```

```python
#Determine the number of unique lemma available in the text.
#basically unique root words

from nltk.stem import WordNetLemmatizer

# Initialize WordNet Lemmatizer
lemmatizer = WordNetLemmatizer()

# Lemmatize the tokens
lemmas = [lemmatizer.lemmatize(token) for token in tokens]

# Find unique lemmas
unique_lemmas = set(lemmas)

# Print the number of unique lemmas
print("Number of unique lemmas:", len(unique_lemmas))
```

```python
from nltk.probability import FreqDist

# Create a frequency distribution of the tokens
fdist = FreqDist(filtered_tokens)

# Get the top 10 most frequent tokens
top_10_tokens = fdist.most_common(10)

# Extract the tokens and their frequencies for plotting
tokens, frequencies = zip(*top_10_tokens)

# Plotting the frequency distribution of the top 10 tokens
plt.figure(figsize=(10, 5))
plt.bar(tokens, frequencies, color='skyblue')
plt.xlabel('Tokens')
plt.ylabel('Frequency')
plt.title('Top 10 Most Frequent Tokens')
plt.xticks(rotation=45)
plt.show()
```

```python
#Find the number of unique bigrams, trigrams and quadgrams (n=4) in the corpus
from nltk.util import ngrams


# Tokenize the text
tokens = word_tokenize(text)

# Generate bigrams, trigrams, and quadgrams
bigrams_list = list(ngrams(tokens, 2))
trigrams_list = list(ngrams(tokens, 3))
quadgrams_list = list(ngrams(tokens, 4))

# Find unique bigrams, trigrams, and quadgrams
unique_bigrams = set(bigrams_list)
unique_trigrams = set(trigrams_list)
unique_quadgrams = set(quadgrams_list)

# Print the number of unique bigrams, trigrams, and quadgrams
print("Number of unique bigrams:", len(unique_bigrams))
print("Number of unique trigrams:", len(unique_trigrams))
print("Number of unique quadgrams:", len(unique_quadgrams))

```

```python
#Find all dates and convert them to the DD-MM-YYYY format.
import re
from datetime import datetime

# Regex pattern to find dates (e.g., 27-March-2000)
date_pattern = r'(\d{1,2})[- ](January|February|March|April|May|June|July|August|September|October|November|December)[- ](\d{4})'

# Function to convert to DD-MM-YYYY format
def convert_to_ddmmyyyy(match):
    day = match.group(1)
    month_str = match.group(2)
    year = match.group(3)
    
    # Convert month name to month number
    month_num = datetime.strptime(month_str, "%B").month
    
    # Format to DD-MM-YYYY
    formatted_date = f"{int(day):02d}-{int(month_num):02d}-{year}"
    return formatted_date

# Replace the dates in the text with formatted dates
formatted_text = re.sub(date_pattern, convert_to_ddmmyyyy, text)

# Find all converted dates
converted_dates = re.findall(r'\d{2}-\d{2}-\d{4}', formatted_text)

# Display results
print("Converted Dates:", converted_dates)
print("\nFormatted Text:\n", formatted_text)
```

```python
#Plot a distribution of the different values of year occurring in the text.

# Regex pattern to find years (assuming years from 1900 to 2099)
year_pattern = r'\b(19[0-9]{2}|20[0-9]{2})\b'

# Extract years from the text
years = re.findall(year_pattern, text)

# Count frequency of each year
year_counts = Counter(years)

# Plotting the distribution
plt.figure(figsize=(8, 5))
plt.bar(year_counts.keys(), year_counts.values(), color='skyblue')
plt.xlabel('Year')
plt.ylabel('Frequency')
plt.title('Distribution of Years in the Text')
plt.xticks(rotation=45)
plt.show()
```

```python
# #Determine whether the text contains any phone numbers in it. Ensure the phone
# numbers are valid if any. If an invalid phone number is found, remove it from
# the text.


# Regular expression to find phone numbers (valid format: +91-XXXXXXXXXX)
phone_pattern = r'(\+91-\d{10})'

# Extract phone numbers from the text
phone_numbers = re.findall(phone_pattern, text)
print(f"Phone numbers found: {phone_numbers}")

# Function to validate phone numbers (for this case, we assume +91-XXXXXXXXXX is valid)
def is_valid_phone(phone):
    return re.fullmatch(r'\+91-\d{10}', phone) is not None

# Check and remove invalid phone numbers
valid_phones = [phone for phone in phone_numbers if is_valid_phone(phone)]

# Remove invalid phone numbers from the text
cleaned_text = text
for phone in phone_numbers:
    if not is_valid_phone(phone):
        cleaned_text = cleaned_text.replace(phone, '')

print("\nValid phone numbers:", valid_phones)
print("\nCleaned text:\n", cleaned_text)
```
